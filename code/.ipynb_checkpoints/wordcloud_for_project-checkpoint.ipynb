```python

koParser = nltk.RegexpParser("""
    NP: {<N.*>*<N.*>?}
    OP: {<N.*>*<JKO>*}
    VP: {<M.*><V.*>*<E.*>*}
""")
```

```python

def parsing_contents (data) : 
    dics = {}
    errorid =[]
    for i in range(len(spec_up)):
        lists = []
        try: #활동 내용 없으면 에러나서 
            koTags1 = kkma.pos(data["활동내용"][i])
            koTags2 = kkma.pos(data["제목"][i])
            tree1 = koParser.parse(koTags1)
            tree2 = koParser.parse(koTags2)
        except : 
            pass
        for subtree1 in tree1.subtrees():
            if subtree1.label() == "NP":
                a = " ".join(e[0] for e in list(subtree1))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)

        for subtree2 in tree2.subtrees():
            if subtree2.label() == "NP":
                a = " ".join(e[0] for e in list(subtree2))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)
                dics[i] = lists
    return [dics , errorid]
```

```python
from konlpy.corpus import kolaw
from collections import Counter
import pytagcloud
```

```python
wordss = []
for i in contents[0]:
     wordss.extend(contents[0][i])
```

```python
for_cloud=[]
for k in wordss:
#     print(len(k))
    if len(k) < 2 or k.isdigit() or k[0].isdigit() or k == '봉사' or k== '자원' or k=='활동' or k=='모집':
        pass
    else : 
        for_cloud.append(k)
```

```python
temp = pd.DataFrame(for_cloud)
```
