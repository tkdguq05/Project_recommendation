```python
from konlpy.tag import Kkma
import nltk
import pandas as pd
```

```python
kkma = Kkma()
```

```python
spec_up = pd.read_csv("mobile_spec_up4.csv",encoding = "ansi")
```

```python

koParser = nltk.RegexpParser("""
    NP: {<N.*>*<N.*>?}
    OP: {<N.*>*<JKO>*}
    VP: {<M.*><V.*>*<E.*>*}
""")
```

```python
spec_up.columns
```

```python

def parsing(data) : 
    dics = {}
    errorid =[]
    for i in range(0,100):
        lists = []
        try: #활동 내용 없으면 에러나서 
            koTags = kkma.pos(data["활동내용"][i])
            tree = koParser.parse(koTags)
        except : 
            errorid.append(i)
        for subtree in tree.subtrees():
            if subtree.label() == "NP":
                a = " ".join(e[0] for e in list(subtree))
                #print(a)

                if a.find(' ') != -1:
                    for j in range(len(a.split())):
                        lists.append(a.split()[j])
                else :
                    lists.append(a)
                dics[i] = lists
    return [dics , errorid]
```

```python
def sum_ma(Matrix):
    sum_ma_li = []
    for i in Matrix :
        sum_ma = 0
        for j in i :
            sum_ma += j
        sum_ma_li.append(sum_ma)
    return sum_ma_li
```

```python
def classli(voca, Matrix) :
    key_voca = list(voca.keys())
    whole = []
    sum_ma_li = sum_ma(Matrix)


    for i in range(len(Matrix)): #애초에 글자수가 적은애들한테는 분류 관대하게, 글자수 많은 애들한테는 분류 빡빡하게
        class_list = []
        for j in range(len(Matrix[i])):
            if sum_ma_li[i] < 5 :
                if Matrix[i][j] > 0 :
                    class_list.append(key_voca[j])
            elif sum_ma_li[i] < 7 :
                if Matrix[i][j] > 1:
                    class_list.append(key_voca[j])
            elif sum_ma_li[i] < 13 :
                if Matrix[i][j] > 2 :
                    class_list.append(key_voca[j])     
            elif sum_ma_li[i] < 17 :
                if Matrix[i][j] > 3:
                    class_list.append(key_voca[j])
            else :
                if Matrix[i][j] > 5:
                    class_list.append(key_voca[j])
                    
        whole.append(class_list)

    return whole
            
        
    
    
```
