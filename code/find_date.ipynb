```python
df = pd.read_csv('outcampus_selenium_re.csv', encoding='ansi')
dueDate = df['모집 기간']
uploadDate = df['작성일']
```

```python
dueDate = dueDate.fillna(0)  # None처리

all_cnt = -1
start = []
end = []

for dt in dueDate:
    all_cnt += 1
#    print(dt)
    totalDate = ''
    if dt == 0:
        start.append('')
        end.append('')
        continue
#    print(dt[0])
    startDate = ''
    endDate = ''

    splitFlag = False
    
    wave_idx = -1
    if '~' in dt:
#        print(dt)
        wave_idx = dt.index('~')
        # 2018xxxx ~ 2018xxxx
        for i in range(len(dt)):
            if str.isdigit(dt[i]): #첫시작점
                if i < wave_idx:
                    splitFlag = True
#                    print(dt)
                    # 10.30~12,30
                break
                
                    
        # ~10.30
    if splitFlag:
#        print(dt)
        started = dt[:wave_idx]
        ended = dt[wave_idx+1:]
        
        startDate = findDate(started)
        endDate = findDate(ended)


#        print(dt)
#        print('started: ',startDate)
#        print('ended: ',endDate)
        
        if not endDate:
#            print(dt)
            if ended[0].isdigit():
                if ended[1].isdigit():
                    endDate = startDate[:6]+ended[:2]
                elif ended[1] == '일':
                    endDate = startDate[:6]+'0'+ended[0]
            elif ended[0] == ' ':
                if ended[1].isdigit():
                    if len(ended) < 2:
                        pass
                    if ended[2].isdigit():
                        endDate = startDate[:6]+ended[1:3]
                    elif ended[2] == '일':
                        endDate = startDate[:6]+'0'+ended[1]
            #print('new endDate = ',endDate)
            else:  # 마감일만 써있는 경우
                endDate = startDate
                startDate = uploadDate[all_cnt]  #게시 날짜로

        start.append(startDate)
        end.append(endDate)

                
    else:
        startDate = uploadDate[all_cnt]
        endDate = findDate(dt)

        start.append(startDate)
        end.append(endDate)

#        print(dt, '-> ', endDate)

df['모집 시작일'] = start
df['모집 마감일'] = end
df = df.drop(['모집 기간'], 1)
df.to_csv('selenium_outcampus_modified_date.csv', encoding='ansi')
```
