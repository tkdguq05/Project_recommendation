```python
age_key_xpath ='//*[@id="frm"]/div[2]/table/tbody/tr[8]/th[2]'
age_value_xpath = '//*[@id="frm"]/div[2]/table/tbody/tr[8]/td[2]/div'
```

```python
driver.find_element_by_xpath('//*[@id="frm"]/div[2]/table/tbody/tr[1]/td[4]').text
```

```python
#모집기간
recruit='//*[@id="frm"]/div[2]/table/tbody/tr[1]/td[4]'
```

```python
#활동장소
location='//*[@id="frm"]/div[2]/table/tbody/tr[5]/td[2]/div/span'
```

```python
jlinks = []
full_links = []
driver = webdriver.Chrome()
data = {}
f = open('culture_edit.csv','w', encoding='cp949')
#wr = csv.writer(f)
f.write('모집기간, 제목, 구분, 활동영역, 활동처명, 활동기간, 모집인원, 봉사내용, 첨부파일, 상해보험, 기타, 활동장소 ,모집연령\n')
for pageIndex in range(1,34):
    full_link = trial+str(pageIndex)
    full_links.append(full_link)
    driver.get(full_links[pageIndex-1])
    print(pageIndex,'-------------------------------')
    print(full_link)

    
    for contentIndex in range(1,10):
        try :
            f.write(driver.find_element_by_xpath(recruit).text)
            f.write(',')
            driver.find_element_by_xpath(contents.format(contentIndex)).click()
        except :
            driver.find_element_by_xpath(contents.format(contentIndex)).send_keys(Keys.ENTER)
            
        
        for bodyIndex in range(0,12):
            if bodyIndex == 0:
                title = driver.find_element_by_xpath('//*[@id="frm"]/div[1]/h2').text
                title_edit = title.replace(',','')
                f.write(title_edit)
                f.write(',')
                continue

            if bodyIndex ==6:
                vol_content = driver.find_element_by_xpath(td_xpath.format(bodyIndex)).text
                preproc = vol_content.replace('\n','')
                edited = preproc.replace(',','')
                f.write(edited)
                f.write(',')
                continue
                
            if bodyIndex == 9 :
                etc = driver.find_element_by_xpath(td_xpath.format(bodyIndex)).text
                pre_etc = etc.replace(',','')
                f.write(pre_etc)
                f.write(',')
                continue
                
            if bodyIndex == 10:
                loc = driver.find_element_by_xpath(location).text
                pre_loc = loc.replace(',','')
                f.write(pre_loc)
                f.write(',')
                continue
                

            
            if bodyIndex == 11:
#                 print('{0} : {1}'.format(last_key, last_value))
                proc = driver.find_element_by_xpath(age_value_xpath).text
                age = proc.replace(',',' ')
                f.write(age)
                f.write(',')
                f.write('\n')
                driver.back()
                continue
            f.write(driver.find_element_by_xpath(td_xpath.format(bodyIndex)).text)
            f.write(',')
        
                        #key=driver.find_element_by_xpath(th_xpath.format(bodyIndex)).text
            #value=driver.find_element_by_xpath(td_xpath.format(bodyIndex)).text
            #last_key=driver.find_element_by_xpath(age_key_xpath).text
            #last_value=driver.find_element_by_xpath(age_value_xpath)
            #print(driver.find_element_by_xpath(td_xpath.format(bodyIndex)).text)
```
